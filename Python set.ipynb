**Built_In_Functions in Python set**

**adding**
- `add`
- `update `

**decrease**
- `pop`
- `discard `
- `remove `

**some common operations**
- `clear `
- `copy `
- `union `
- `intersection `

```python
# add=this function will take one value from user and add randomlly somewhere

a = {10,20,30,50}

a.add(600)

print(a)
```

```python
# update=will add multiple values randomlly

a = {10,20,30,50}

a.update([100,200,300])

print(a)
```

```python
# pop = will remove one value randomlly

a = {10,20,30,50}

a.pop()

print(a)
```

```python
# discard = if the value is there it will remove
# or else it will be silent

a = {10,20,30,50}

a.discard(30)

print(a)
```

```python
# remove if the value is there it will remove
# or else it will through error

a = {10,20,30,50}

a.remove(20)

print(a)
```

```python
# discard = if the value is there it will remove
# or else it will be silent

a = {10,20,30,50}

a.discard(300)

print(a)
```

**From Above operations Python set is Mutable**

```python
a = {1,2,3,4,5,6}
b = {5,6,7,8,9,10}

print(a.union(b))
```

```python
# intersection

a = {1,2,3,4,5,6}
b = {5,6,7,8,9,10}

print(a.intersection(b))
```

```python
# copy
a = {10,20,30,40}
b = a # deep copy

print(a)
print(b)

print(id(a))
print(id(b))

a.add(100)

print(a)
print(b)
```

```python
# copy
a = {10,20,30,40}
b = a.copy() # shallow copy

print(a)
print(b)

print(id(a))
print(id(b))

a.add(100)

print(a)
print(b)
```

```python
# clear

a = [10,20,30] # save in Memory

a.clear()

print(a)
```

```python
# clear

a = {10,20,30} # save in Memory

a.clear()

print(a)
```
