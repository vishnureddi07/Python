**String in Python is Mutable or Immutable**

- `Mutable -> means we can increase and decrease the Memory size `
- `immutable -> means we can't increase and decrease the Memory size `

**Strings in python is Mutable or Immtable -> prove it**

- `try to increase the memory size `
- `try to decrease the memory size `


```python
# trying to increase the Memory size:

a = 'data' # save in the Memory

a[4] = 'j' # trying to fill 'j' in 4th box



print(a) # calling
```

```python
# trying to decrease

a = 'data' # save in Memory

del a[2]

print(a) # calling
```

**From above operations we can say that Python strings are Immutable**

```python
# capitalize=will convert starting letter into capital and remaining all are small

a = 'pARIS uSSR' # save in the Memory

a = a.capitalize()

print(a) # calling from Memory
```

```python
# title=each word first letter into capital remaining all small

a = 'uSSr paRIs' # save in the Memory

a = a.title()

print(a) # calling
```

```python
# lower=will convert capital letter into small letter

a = 'paRIsss' # save

print(a.lower())
```

```python
# upper=will convert small letter into capital

a = 'paRIsss' # save

print(a.upper())
```

```python
# index = we can find the index of a character

f = 'data science' # save in the Memory

print(f.index('i'))
```

```python
# index = we can find the index of a character

f = 'data science' # save in the Memory

print(f.index('a'))
```

```python
# index = we can find the index of a character

f = 'data science' # save in the Memory

print(f.index('a' , 2))
```

```python
# index = we can find the index of a character

f = 'data science' # save in the Memory

print(f.index('e',9))
```

```python
# count=is used to count the number of occurences

f = 'data science' # save in the Memory

print(f.count('e'))
```

```python
# replace

f = 'data' # save in the Memory

f = f.replace('t' , 'j')

print(f)
```

```python
# replace

f = 'data' # save in the Memory

f = f.replace('a' , 'j' , 1)

print(f)
```

```python
# islower = generate boolean outputs

a = 'paris'

print(a.islower())
```

```python
# isupper = generates boolean outputs
a = 'PARIS'

print(a.isupper())
```

```python
# isalpha = all are alphabets or not

a = 'usa'

print(a.isalpha())
```

```python
# isnumeric = all are numbers or not

a = '156'

print(a.isnumeric())
```

```python
# isalnum = we can have both numbers and alphabets and combination also

f = 'kamal123'

print(f.isalnum())
```

```python
# startswith =  boolean outcome

p = 'india is powerfull'

print(p.startswith('i'))
```

```python
# endswith = boolen outcome

p = 'india is powerfull'

print(p.endswith('l'))
```

```python
# membership concept = checking substring present in main string or not:

t = 'i love programming language' # main string

print('love' not in t) # 'love' -> substring
```
