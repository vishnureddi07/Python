**Python Dictionary**

- `Python Dictionary concept is represented with -> {}`
- `Anything if we pass inside the {} -> it will become set but if we pass key_value pair inside the {} -> then it is a dictionary `


- `where key -> column name and value means -> data inside the column `

- `Dictionary concept is used to prepare the data but we can see the exsheet -> but if we want give the dictionary data to pandas library `

- `In dictionary we dont have the index concept -> so if there is no index | slicing concept and skipping concept is not possible `

```python
f = {'Name':'John' , 'Age':27, 'Salary' : 1200}
print(f)
```

```python
f = {'Name':['John','peter','brad','smith'],
     'Age':[27,30,35,40],
     'Salary':[1200,1400,2000,1000]
     }

print(f)
```

**Python Dictionary concept is Mutable or Immutable -> Prove it**

- `try to increase the Memory size `
- `try to decrease the Memory size `

```python
fruits = {'apple':100 ,'banana':200 ,'carrot':300}

fruits['apple'] = 500

print(fruits)
```

```python
fruits = {'apple':100 ,'banana':200 ,'carrot':300}

fruits['grapes'] = 500

print(fruits)
```

```python
fruits = {'apple':100 ,'banana':200 ,'carrot':300}

del fruits['banana']

print(fruits)
```

**Built_in_Functions**

- `get`
- `update `
- `pop `
- `popitem `
- `clear `
- `copy `
- `keys `
- `values `

```python
# get = to access values

fruits = {'apple':100 ,'banana':200 ,'carrot':300}

print(fruits['banana'])
```

```python
fruits = {'apple':100 ,'banana':200 ,'carrot':300}

print(fruits.get('banana'))
```

```python
# update:

fruits = {'apple':100,'banana':200,'carrot':300}

fruits['apple'] = 500

print(fruits)
```

```python
# update:

fruits = {'apple':100,'banana':200,'carrot':300}

fruits.update({'apple':500})

print(fruits)
```

```python
# update:

fruits = {'apple':100,'banana':200,'carrot':300}

fruits['grapes'] = 500

print(fruits)
```

```python
# update:

fruits = {'apple':100,'banana':200,'carrot':300}

fruits.update({'grapes':500})

print(fruits)
```

```python
fruits = {'apple':100 ,'banana':200 ,'carrot':300}

del fruits['banana']

print(fruits)
```

```python
# pop = it will take key and removes

fruits = {'apple':100 ,'banana':200 ,'carrot':300}

fruits.pop('banana')

print(fruits)
```

```python
# popitem = will remove last key

fruits = {'apple':100 ,'banana':200 ,'carrot':300}

fruits.popitem()

print(fruits)
```

```python
# keys

fruits = {'apple':100 ,'banana':200 ,'carrot':300}

print(fruits.keys())
```

```python
# values

fruits = {'apple':100 ,'banana':200 ,'carrot':300}

print(fruits.values())
```

```python
# items

fruits = {'apple':100 ,'banana':200 ,'carrot':300}

print(fruits.items())
```

```python
# copy

fruits = {'apple':100 ,'banana':200 ,'carrot':300}
b = fruits
print(fruits)
print(b)
print(id(fruits))
print(id(b))
fruits['apple'] = 500
print(fruits)
print(b)
```

```python
# copy

fruits = {'apple':100 ,'banana':200 ,'carrot':300}
b = fruits.copy()
print(fruits)
print(b)
print(id(fruits))
print(id(b))
fruits['apple'] = 500
print(fruits)
print(b)
```

```python
# clear
fruits = {'apple':100 ,'banana':200 ,'carrot':300}

fruits.clear()

print(fruits)
```
