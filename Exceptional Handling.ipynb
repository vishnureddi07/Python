**Exceptional Handling**

`works with try block and except block `

```python
a = 10
b = 20

print(a + b)
```

```python
try:
  a = 10
  b = 20
  print(a + b)


except Exception as e:
  print(e)
```

```python
# finally in exceptional handling will run with error
# without error also

try:
  a = 10
  b = 20
  print(a / b)

except Exception as f:
  print(f)

finally:
  print('i will execute without error')
```

```python
import sys
```

```python
try:
  a = 10
  b = 20
  print(a / b)

except Exception as f:
  print(f'Error Msg by try block : {f}')
```

```python
try:
  a = 10
  b = 0
  print(a / b)

except Exception as f:
  print(f'Error Msg by try block : {f}')
  er_type,er_msg,er_line = sys.exc_info()
  print(f'Error type : {er_type}')
  print(f'Error Msg by sys : {er_msg}')
  print(f'Line No : {er_line.tb_lineno}')
```
