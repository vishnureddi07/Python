**Rules to create a variables**

- `Rule No:1 -> Varibale should not start with a Number -> if you want to start with a Number Please Use Underscore `

- `Rule No:2 -> Varibale should be single character or single word multiple characaters or Multiple words are not allowed if you want Please use underscore `

- `Rule No:3 -> Python Varibales are Case Sensitive (case sensitive -> small letter and capital letters )`

- `Rule No:4 -> We should not use below Punctuations as a varibale -> !"#$%&'()*+,-./:;<=>?@[\]^_`{|}~`

```python
10a = 25 # save in the Memory

print(10a) # calling from the Memory
```

```python
_10a = 25 # save in the Memory

print(_10a) # calling from the Memory
```
