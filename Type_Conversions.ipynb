**Type Conversions**

`Type Conversions Means -> changing the data type into other data types:`

- `int -> float and int -> str `
- `float -> int and float -> str `
- `str -> int and str -> float `

```python
# int -> float

a = 24 # save in the Memory
print(a) # calling from memory
print(type(a)) # checking the data type

b = float(a) # actual conversions is going on here

print(b) # calling from the Memory
print(type(b)) # checking the data type:
```

```python
a = 24
b = str(a)
print(b)
```

```python
# int -> str

a = 24 # save in the Memory
print(a) # calling from memory
print(type(a)) # checking the data type

b = str(a) # actual conversions is going on here

print(b) # calling from the Memory
print(type(b)) # checking the data type:
```

```python
# float -> int

a = 24.55 # save in the Memory
print(a) # calling from memory
print(type(a)) # checking the data type

b = int(a) # actual conversions is going on here

print(b) # calling from the Memory
print(type(b)) # checking the data type:
```
