**Python Comprehensions -> are used to reduce the level of code**

In comprehensions we have 2 types:

- `List Comprehension`
- `Dictionary Comprehension`

**List Comprehension**

- `syntax_1 = one forloop and one print condition `

     `syntax = [<print_ara> <forloop>]`


- `syntax_2 = one forloop and one if condition and one if print `

     `syntax = [<print_area> <forloop> <if_condition>]`

- `syntax_3 = one forloop one if and if print and one else and else print `

      `syntax = [<if_print> <if_condition> else <else_print> <forloop>]`

```python
for i in range(1,6):
  print('Good Morning')
```

```python
print(['Good Morning' for i in range(1,6)])
```

```python
for i in range(11,16):
  if i % 2 == 0:
    print(f'Even : {i}')
```

```python
print([f'Even : {i}' for i in range(11,16) if i % 2 == 0])
```

```python
for i in range(11,16):
  if i % 2 == 0:
    print(f'Even : {i}')
  else:
    print(f'Odd : {i}')
```

```python
print([f'Even : {i}' if i % 2 == 0 else f'Odd : {i}' for i in range(11,16)])
```

**Dictionary Comprehensions**

```python
fruits = {'apple':100 , 'banana':200 , 'carrot':300}

for i in fruits:
  fruits[i] = fruits[i] * 2

print(fruits)
```

```python
fruits = {'apple':100 , 'banana':200 , 'carrot':300}

print({i:fruits[i]*2 for i in fruits})
```

```python
fruits = {'apple':100 , 'banana':200 , 'carrot':300}

d = {}

for i in fruits:
  if fruits[i] > 150:
    d[i] = fruits[i]

print(d)
```

```python
fruits = {'apple':100 , 'banana':200 , 'carrot':300}

print({i:fruits[i] for i in fruits if fruits[i] > 150})
```

```python
fruits = {'apple':100 , 'banana':200 , 'carrot':300}

for i in fruits:
  if fruits[i] > 150:
    fruits[i] = fruits[i] * 3
  else:
    fruits[i] = fruits[i] * 2

print(fruits)
```

```python
fruits = {'apple':100 , 'banana':200 , 'carrot':300}

fruits.keys()
```

```python
fruits = {'apple':100 , 'banana':200 , 'carrot':300}

fruits.values()
```

```python
fruits = {'apple':100 , 'banana':200 , 'carrot':300}

fruits.items()
```

```python
print({i: (j * 3 if j > 150 else j * 2) for i,j in fruits.items()})
```

```python
import time
```

```python
starting_time = time.time()

fruits = {'apple':100 , 'banana':200 , 'carrot':300}
for i in fruits:
  if fruits[i] > 150:
    fruits[i] = fruits[i] * 3
  else:
    fruits[i] = fruits[i] * 2


print(f'Total time to Execute : {time.time() - starting_time}')
```

```python
starting_time = time.time()

fruits = {'apple':100 , 'banana':200 , 'carrot':300}

print({i: (j * 3 if j > 150 else j * 2) for i,j in fruits.items()})

print(f'Total time to Execute : {time.time() - starting_time}')
```
