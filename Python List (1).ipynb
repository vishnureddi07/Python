**Python List concept is Mutable or Immutable -> Prove it**

- `try to increase the Memory size `
- `try to decrease the Memory size `


```python
p = 'data' # save in the Memory

p[4] = 'j'

print(p) # calling
```

```python
# try to increase the Memory size

f = [10,50,90] # save in the Memory

f[3] = 100

print(f) # calling
```

- `in python string concept adding and decreasing is not at all possibe so its a immutable `

- `in python list concept we can add and we can decrease the memory size but manually not possible -> we need to take help of list built in functions`


**Python List Built_in_functions**


**Increase**

- `append `
- `extend `
- `insert `

**decrease**
- `pop `
- `remove `

**some common operations**
- `index `
- `count `
- `reverse`
- `sort `
- `clear `
- `copy `

```python
# append=this function will take only one input from user and will add to the end of the existing list

f = [10,50,90] # save in the Memory

f.append(100)

print(f) # calling
```

```python
# extend=this function will take multiple inputs from user and will add to the end of the existing list

f = [10,50,90] # save in the Memory

f.extend([100,200,500])

print(f) # calling
```

```python
# insert = we can add value where we want

f = [10,50,90] # save in Memory

f.insert(1 , 100)

print(f) # calling
```

```python
# pop=this function defaulty will remove last value
# but if we pass input it will remove that index value

f = [10,50,90] # save in the Memory

f.pop()

print(f) # calling
```

```python
# pop=this function defaulty will remove last value
# but if we pass input it will remove that index value

f = [10,50,90] # save in the Memory

f.pop(1)

print(f) # calling
```

```python
# remove = will take value from user and will remove that value:

f = [10,50,90] # save in the Memory

f.remove(10)

print(f) # calling
```

```python
# remove = will take value from user and will remove that value:

f = [10,50,90] # save in the Memory

f.remove(100)

print(f) # calling
```

**from the above operations we can say that Python List concept is Mutable**

```python
# index

a = [10,20,30,50,20] # save in the Memory

print(a.index(20 , 2))
```

```python
# count

a = [10,20,30,50,20] # save in the Memory

print(a.count(20))
```

```python
# reverse -> [::-1]

a = [10,20,30,50,90] # save in the Memory

a.reverse()

print(a) # calling
```

```python
# sort -> ascending order and descending order

f = [10,5,19,21,6] #save in Memory

f.sort() # ascending order

print(f) # calling
```

```python
# sort -> ascending order and descending order

f = [10,5,19,21,6] #save in Memory

f.sort(reverse=True) # Descending order

print(f) # calling
```

```python
g = [10,20,30] # save in the Memory

del g[2]

print(g) # caling
```

```python
g = [10,20,30] # save in the Memory

del g

print(g) # calling
```

```python
# clear=will remove values from varibale and remains empty
# but it will not remvoe varibale

g = [10,20,30] # save in the Memory

g.clear()

print(g) # calling
```

```python
# copy=in copy we have 2 types
# deep copy
# shallow copy
```

```python
a = [10,20,30]
b = a # deep copy

print(a)
print(b)

print(id(a))
print(id(b))

a.append(100)

print(a)
print(b)
```

```python
a = [10,20,30]
b = a.copy() # shallow copy

print(a)
print(b)

print(id(a))
print(id(b))

a.append(100)

print(a)
print(b)
```
