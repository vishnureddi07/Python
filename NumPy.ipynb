```python
# array -> is used to develop 1D or 2D or 3D etc

a = np.array([10,20,50,90,45])

print(a) # printing 1D values

print(a.shape) # (rows and columns)
print(a.size) # total values

print(a.ndim) # tells how many dimensions
```

```python
print(a[2])
```

```python
# 2D

a = np.array([  [11,4,17]  , [2,8,3] , [5,7,1] ])

print(a) # printing 1D values

print(a.shape) # (rows and columns)
print(a.size) # total values

print(a.ndim) # tells how many dimensions
```

```python
a[1][2]
```

```python
a[2][0]
```

```python
# arange = is used to create only 1D data with (n-1) logic

b = np.arange(11,16)
print(b)
print(b.ndim)
```

```python
b = np.arange(11,19)
print(b)
print(b.ndim)

print('---------------')

c = b.reshape(2,4)
print(c)
print(c.shape)
```

```python
# 1D to 2D

b = np.arange(1,61).reshape(10,6)
print(b)
print(b.shape)
print(b.size)
```

```python
b[4][3]
```

```python
# 1D to 3D

a = np.arange(1,61).reshape(2,5,6)
a
```

```python
a[1][2][4]
```

```python
a[2][1][2]
```

`reshape function you can convert 1D to 2D -> 2D to 3D anything is possible `

`But reshape(-1,1) and reshape(1,-1) is used to convert 1D to 2D`

```python
# reshape(-1,1) and reshape(1,-1)

a = np.array([10,20,30,50,90,70])
print(a)
print(a.ndim)
print(a.shape)

print('----------------')
b = a.reshape(-1,1)
print(b)
print(b.shape)
print(b.ndim)
```

```python
# reshape(-1,1) and reshape(1,-1)

a = np.array([10,20,30,50,90,70])
print(a)
print(a.ndim)
print(a.shape)

print('----------------')
b = a.reshape(1,-1)
print(b)
print(b.shape)
print(b.ndim)
```

```python
# ravel = is used to convert any dimension into 1D
a = np.array([  [1,10,5] , [7,6,11] , [5,3,2] ])
print(a)
print(a.ndim)

b = np.ravel(a)
print(b)
print(b.shape)
```

```python
# Transpose = convert rows to columns and columns to rows

a = np.array([  [1,10,5] , [7,6,11] , [5,3,2] ])
print(a)

print('-------------')
b = np.transpose(a)
print(b)
```

```python
# concatination = merging the data

a = np.array([ [10,11] , [12,13] ])
b = np.array([ [100,1] , [200,2] ])

c = np.concatenate((a, b))
print(c)
```

```python
# matmul

a = np.array( [ [1,2] , [4,3] ])
b = np.array([ [1,4,9] , [3,6,2] ])

np.matmul(a,b)
```

```python
b = np.array([ [1,4,9] , [3,6,2] , [4,6,2] ])
b
```

```python
for i in b:
  for j in i:
    if j % 2 != 0:
      print(j)
```

```python
r = 0

for i in b:
  c = 0
  for j in i:
    if j % 2 != 0:
      print(j)
    c = c + 1

  r = r + 1

print(f'Number of rows : {r}')
print(f'Number of columns : {c}')
```
