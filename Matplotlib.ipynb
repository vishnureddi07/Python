```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

`Matplotlib.pyplot -> Built_in_Functions`

- `Scatter Plot `
- `Line Chart `
- `Legend Function `
- `subplots `
- `Histogram `
- `Bar Chart `
- `Pie_chart `

```python
# scatter plot

india = [1,4,7,3,6,2]
usa = [4,6,2,1,5,7]

plt.scatter(x = india , y = usa)
```

```python
# scatter plot

india = [1,4,7,3,6,2]
usa = [4,6,2,1,5,7]

plt.figure(figsize=(5,2))
plt.title('Scatter plot')
plt.xlabel('Indian_values')
plt.ylabel('USA_values')

plt.scatter(india,usa,color = 'g',marker='*')

plt.show()
```

```python
# line chart

india = [1,4,7,3,6,2]
usa = [4,6,2,1,5,7]

plt.figure(figsize=(5,2))
plt.title('Line plot')
plt.xlabel('Indian_values')
plt.ylabel('USA_values')

plt.plot(india,usa,color = 'g',marker='*')

plt.show()
```

```python
india = [1,4,7,3,6,2]
usa = [4,6,2,1,5,7]
uk = [5,2,1,6,3,4]

plt.figure(figsize=(5,2))
plt.title('Line plot')


plt.plot(india,usa)
plt.plot(india,uk)

plt.show()
```

```python
india = [1,4,7,3,6,2]
usa = [4,6,2,1,5,7]
uk = [5,2,1,6,3,4]

plt.figure(figsize=(5,2))
plt.title('Line plot')


plt.plot(india,usa,label = 'ind-usa')
plt.plot(india,uk,label = 'india-uk')

plt.legend(loc = 1)
plt.show()
```

```python
#subplot

plt.figure(figsize=(7,2))

plt.subplot(1,2,1)
plt.title('India-USA')
plt.plot(india,usa,marker='*',color='r')

plt.subplot(1,2,2)
plt.title('india-uk')
plt.plot(india,uk,color='g',marker='*')

plt.show()
```

```python
#subplot

plt.figure(figsize=(3,5))

plt.subplot(2,1,1)
plt.title('India-USA')
plt.plot(india,usa,marker='*',color='r')

plt.subplot(2,1,2)
plt.title('india-uk')
plt.plot(india,uk,color='g',marker='*')

plt.show()
```

```python
# Histogram = is used to find the data points in between the Intervals
plt.figure(figsize=(5,2))

age = [1,4,7,9,3,11,14,17,21,22,27,29,32,33]

plt.hist(x = age , bins = 6)
plt.show()
```

```python
# Bar chart
company = ['Google','Meta','IBM','Infosys']
revenue = [1600 , 1900 , 1100 , 800]

plt.figure(figsize=(5,3))
plt.bar(company,revenue,color=['r','g','black','y'])
plt.show()
```

```python
# Bar chart
company = ['Google','Meta','IBM','Infosys']
revenue = [1600 , 1900 , 1100 , 800]

plt.figure(figsize=(5,3))
plt.barh(company,revenue,color=['r','g','black','y'])
plt.show()
```

```python
# Pie chart
company = ['Google','Meta','IBM','Infosys']
revenue = [1600 , 1900 , 800 , 400]

plt.figure(figsize=(5,3))

plt.pie(revenue,labels = company)
plt.show()
```

```python
# Pie chart
company = ['Google','Meta','IBM','Infosys']
revenue = [1600 , 1900 , 800 , 400]

plt.figure(figsize=(5,3))

plt.pie(revenue,labels = company,shadow=True)
plt.show()
```

```python
# Pie chart
company = ['Google','Meta','IBM','Infosys']
revenue = [1600 , 1900 , 800 , 400]

plt.figure(figsize=(5,3))

plt.pie(revenue,labels = company,shadow=True,autopct='%1.2f%%',colors=['olivedrab', 'rosybrown', 'gray', 'saddlebrown'])
plt.show()
```
