```python
class ECE:
  def __init__(self,p,q):
    self.a = p
    self.b = q

  def sharuk(self):
    for i in range(self.a,self.b):
      if i % 2 == 0:
        print(f'Even : {i}')

anil = ECE(1,11)
anil.sharuk()
```

**Inheritance**


`Inheritance concept is known as parent and child concept (relation )`

`In Python we have 5 types of Inheritances`

      - `Single Inheritance `
      - `Multiple Inheritance`
      - `MultiLevel Inheritance `
      - `Hybrid Inheritance `
      - `Hierarichal Inheritance `

```python
# Single Inheritance

class PARENT:
  def building(self):
    print('I have my own Building')


class CHILD(PARENT):
  def job(self):
    print('I have My own Job')


anil = PARENT()
anil.building()
```

```python
# Multiple Inheritance

class FATHER:
  def building(self):
    print('I have my own Building')


class MOTHER:
  def job(self):
    print('I have My own Job')

class CHILD(FATHER,MOTHER):
  def shopping_mall(self):
    print('I am having a MALL')

anil = CHILD()
anil.shopping_mall()
anil.job()
anil.building()
```

```python
# Hierarichal Inhertiance

class FATHER:
  def building(self):
    print('I have my own Building')

class B1(FATHER):
  def job(self):
    print('I have My own Job')

class B2(FATHER):
  def shopping_mall(self):
    print('I am having a MALL')

class B3(FATHER):
  def bussiness(self):
    print('I have my own Bussiness')

anil = B2()
anil.shopping_mall()
anil.building()
```

```python
# Hybrid Inheritance

class GRANDFATHER:
  def building(self):
    print('I have my own Building')

class FATHER(GRANDFATHER):
  def job(self):
    print('I have My own Job')

class MOTHER(GRANDFATHER):
  def shopping_mall(self):
    print('I am having a MALL')

class CHILD(FATHER,MOTHER):
  def bussiness(self):
    print('I have my own Bussiness')

anil = CHILD()
anil.bussiness()
anil.job()
anil.shopping_mall()
anil.building()
```

**Method Overriding**

`Method Overriding concept is only possible when we have one parent class and one child class `

```python
def fun():
  print("good Morning")

def fun():
  print("good Evening")

fun() # calling the function
```

```python
class MANAGER:
  def database_storing(self):
    print('I have a issue here')


class TL(MANAGER):
  def database_storing(self):
    print('I will solve the issue')


anil = TL()
anil.database_storing()
```

```python
def fun(a = None,b = None):
  if a == None and b == None:
    print('I am from USA')
  elif a == None and b!=None:
    print('I am from india')
  elif a != None and b == None:
    print('I am from UK')
  else:
    print('I am happy')

fun()
fun(a = 10)
fun(b = 20)
fun(10,20)
```

**Abstraction**


```python
from abc import ABC,abstractmethod

class PAYTMV1(ABC):
  def flight_booking(self):
    print('I can book flights')

  @abstractmethod
  def gateway(self):
    pass

class PAYTMV2:
  def gatway(self):
    print('Logic here')

obj = PAYTMV2()
obj.gatway()
```
