**what is the difference Between Global Varibales and Local varibales**

```python
# Global varibale = we can access from any where
# we can access from inside the function and outside the function


t = 10 # global varibale

def sharuk():
  print(f'I am Inside the function : {t}')



sharuk() # calling
print(f'I am outside the function : {t}')
```

```python
# local varibale
# we can access only from inside the function directly outside nor possible


def sharuk():
  p = 90 # local variable
  print(f'I am Inside the function : {p}')



sharuk() # calling
print(f'I am outside the function : {p}')
```

```python
# local varibale
# we can access only from inside the function directly outside nor possible


def sharuk():
  global p
  p = 90 # local variable changing to global varibale
  print(f'I am Inside the function : {p}')



sharuk() # calling
print(f'I am outside the function : {p}')
```
