**What is Lambda Function**

- `lambda function is also called as ananymous Function `
- `Ananymous function means -> a Function without a Function name `

- `syntax -> lambda arguments : expressions `


```python
a = 10
b = 20
print(a + b)
```

```python
# Normal functions

def sharuk(a,b):
  print(a+b)



sharuk(10,20) # calling
```

```python
c = lambda a,b : a + b

print(c(10,20))
```

```python
f = lambda x : x * x

print(f(5))
```

```python
n = 2
for i in range(1,11):
  print(f'{n} * {i} = {n * i}')
```

```python
f = lambda a,b : a * b

n = 2
for i in range(1,11):
  print(f'{n} * {i} = {f(n,i)}')
```

**Map | Filter | reduce**

- `map syntax -> map(function_name , iterables)`

```python
def sharuk(b):
  if b % 2 == 0:
    return b



sol = map(sharuk , [10,21,32,47,18])
print(list(sol))
```

```python
def sharuk(b):
  if b % 2 == 0:
    return b



sol = filter(sharuk , [10,21,32,47,18])
print(list(sol))
```

```python
sol = map(lambda x : x * x , [1,5,3,9,7])

print(list(sol))
```

```python
sol = filter(lambda x : x * x , [1,5,3,9,7])

print(list(sol))
```

```python
sol = map(lambda x : x % 2 == 0,[1,4,9,7,6,3])

print(list(sol))
```

```python
sol = filter(lambda x : x % 2 == 0,[1,4,9,7,6,3])

print(list(sol))
```

```python
# reduce
import functools

functools.reduce(lambda x,y : x + y ,[1,4,7,6,3,2])
```

**map with filter**

```python
sol = map(lambda x:x+20,filter(lambda x:x%2==0,[1,4,7,6,9,8,12,13]))
print(list(sol))
```

**Filter with map**

```python
sol = filter(lambda x:x%2==0,map(lambda x:x+20,[1,4,7,6,9,8,12,13]))
print(list(sol))
```

**reduce with map and filter**

```python
sol=functools.reduce(lambda x,y:x+y,map(lambda x:x+20,filter(lambda x:x%2==0,[10,11,12,13,21,22,23,27,29])))
print(sol)
```

**Recurssions**

`A function calling itself `

```python
def sharuk(r):
  if r == 0 or r == 1:
    return 1
  else:
    return r * sharuk(r-1)


sol = sharuk(5)
print(sol)
```
