`File means where we are going to store the data `

`Ex: pdf | txt | word | excel | json etc...`

- `File Handling concept is used to work with any file `
- `suppose if we have a file [we can write data into file | read data from file | add data into the file ]`


`In File Handling concept Mainly we have 3 types (Modes)`

- `w -> write mode`
- `r -> read mode `
- `a -> add mode `

**W -> Write Mode**

- `step_1 : If the File is not available Mode (w) will create the File and will Write the data into the File`

- `step_2 : If the File is available and If the File is Empty Mode (w) will open the File and we can write the data into the file `

- `step_3 : If the File is Available and In the File if we have data and If we pass New data old data will be overridden with New data `

```python
a = open('sharuk.txt' , 'w')

a.write('I am from India')
a.close()
```

```python
a = open('paris.txt' , 'w')

a.write('I am from USA')
a.close()
```

```python
a = open('sharuk.txt' , 'w')

a.write('data science')
a.close()
```

**a -> Add Mode**

- `step_1 : If the File is not available Mode (a) will create the File and will Write the data into the File`

- `step_2 : If the File is available and If the File is Empty Mode (a) will open the File and we can write the data into the file `

- `step_3 : If the File is Available and In the File if we have data and If we pass New data old data will not be overridden with New data but it will add the new data to old data `

```python
a = open('sharuk.txt' , 'a')

a.write('I am from India')
a.close()
```

```python
a = open('paris.txt' , 'a')

a.write('I am from USA')
a.close()
```

```python
a = open('sharuk.txt' , 'a')

a.write(' data science')
a.close()
```

**r_read Mode**

```python
a = open('sharuk.txt','r')

print(a.read())

a.close()
```

`details about close`

```python
a = open('sharuk.txt','w')

a.write('I am from India')
```

```python
with open('sharuk.txt','w') as a:
  a.write('I am from India')
```

```python
a = open('sharuk.txt','r')

print(a.read())

a.close()
```

```python
with open('sharuk.txt','r') as a:
  print(a.read())
```
