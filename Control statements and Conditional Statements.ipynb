**Find the Given Number is palindrome or Not**

```python
n = 121
b = n
s = 0

while n > 0:
  res = n % 10
  s = (s * 10) + res
  n = n // 10

if b == s:
  print(f'{b} : Palindrome')
else:
  print(f'{b} : Not A palindrome')
```

**AMstrong**

```python
n = 153
b = n
s = 0

while n > 0:
  res = n % 10
  s = s + (res * res * res)
  n = n // 10

if b == s:
  print(f'{b} : Amstrong')
else:
  print(f'{b} : Not A Amstrong')
```

**Fibonacci Series**

```python
n = 8

a = 0
b = 1

print(a)
print(b)

for i in range(n-2):
  c = a + b
  print(c)
  a = b
  b = c
```
