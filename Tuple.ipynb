```python
# split = used to break a paragraph into words
# join = used to combine the words into paragraph
```

```python
a = 'usa uk' # save in the Memory

b = a.split()

print(b)
```

```python
a = 'i am from india'
print(a)

b = a.split()
print(b)
```

```python
a = 'i.am.from india'
print(a)

b = a.split('.')
print(b)
```

```python
f = ['usa','uk','paris']

'->'.join(f)
```

**Python Tuples**

- `Python tuple will be represented with -> ()`
- `In Between the () -> we can pass any data type `
- `Memory Allocation | Indexing | slicing | skipping | concepts are same like string and List only -> but Mutable and Immtable and built_in_functions are different `

```python
a = (10,25.55,'data',False) # save in the Memory

print(a) # calling
```

```python
print(a[1])
```

```python
print(a[-1])
```

```python
print(a[2])
```

```python
print(a[2][2])
```

```python
print(a[2 :])
```

```python
print(a[ : 3])
```

```python
print(a[ : ])
```

```python
print(a[: : 2])
```

**Python tuple is Mutable or Immutable -> Prove it**

- `try to increase the Memory size `
- `try to decrease the Memory size `

```python
# trying to increase

a = (10,20,30) # save in Memory

a[3] = 50

print(a)
```

```python
# trying to decrease

a = (10,20,30) # save in Memory

del a[1]

print(a)
```

**From the above operations we can say that python tuple concept is Immutable**

Only 2 built in functions in tuple

- `Index `
- `count `

```python
a = (10,20,50,90,17,12)

print(a.index(90))
```

```python
a = (10,20,50,90,17,12)

print(a.count(50))
```
